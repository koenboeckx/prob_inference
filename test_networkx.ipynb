```python
import networkx as nx
G = nx.Graph()
```

```python
# adding nodes
G.add_node(1)
G.add_nodes_from([2, 3])
G.add_nodes_from([
    (4, {"color": "red"}),
    (5, {"color": "green"}),
])
```

```python
# adding edges
G.add_edge(1, 2)
e = (2, 3)
G.add_edge(*e)
```

```python
DG = nx.DiGraph()
DG.add_weighted_edges_from([(1, 2, 0.5), (3, 1, 0.75)])
DG.out_degree(1, weight='weight')

DG.degree(1, weight='weight')

list(DG.successors(1))

list(DG.neighbors(1))
```
